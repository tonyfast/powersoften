```python
    from articles import *; from pandas import *; from bokeh import plotting, resources
    plotting.output_notebook(resources=resources.CDN)
```

```python
    topics = 'design, systems, materials, community, media, culture'.replace(' ', '').split(',')
```

```python
    nodes = [[-1,0], [-.5, .5], [0, 1], [.5, .5], [1,0], [0,0], [-1, 0]]
    nodes = DataFrame(nodes, columns=list('xy'))
    inner = nodes.iloc[list(range(1, 7, 2))+[1]]
```

```python
    edges = nodes.rolling(2).mean().dropna().set_index(Index(topics))
```

```python
    p = plotting.figure()
    p.line(nodes.x, nodes.y)
    p.line(inner.x, inner.y)
    p.text(edges.x, edges.y, edges.index)
    plotting.show(p)
```
